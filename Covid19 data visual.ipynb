```python
df2 = pd.read_csv("/content/covid_vaccine_statewise.csv")
df2.head()
```

```python
df2.info()
```

```python
df2 = df2.rename(columns= {'Updated On':'Date','Total Doses Administered':'TotalDoses','Male(Individuals Vaccinated)':'Male','Female(Individuals Vaccinated)':'Female', 'Total Individuals Vaccinated':'TotalVaccinated',' Covaxin (Doses Administered)':'Covaxin','CoviShield (Doses Administered)':'CoviShield','Sputnik V (Doses Administered)':'Sputnik'})
df2.Date = pd.to_datetime(df2.Date, format="%d/%m/%Y")
```

```python
df2= df2.groupby(['State'])["TotalDoses" , "Sputnik" , "Covaxin" , "CoviShield" , "TotalVaccinated"].sum().reset_index()
```

```python
df2.dropna(inplace = True)
```

```python
df2 = df2[df2['State'] !='India']
```

```python
fig = px.bar(df2, x='State', y='TotalDoses',text='TotalDoses' , color ="TotalDoses")
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
# fig.update_layout(width=1000, height=1000, bargap=0.05)
fig.show()
```

```python
fig = px.bar(df2 , x = "State" , y=["Covaxin" , "CoviShield"])
fig.show()
```

```python
fig = px.pie(df2 , values = "TotalVaccinated" , names="State" , color_discrete_sequence=px.colors.sequential.Viridis)
fig.update_layout(width=800, height=800)
fig.show()
```
